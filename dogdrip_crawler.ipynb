```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup
import time
import csv
import sys
import io

# sys.stdout = io.TextIOWrapper(sys.stdout.detach(), encoding = 'utf-8')
# sys.stdout = io.TextIOWrapper(sys.stderr.detach(), encoding = 'utf-8')

f = open('write4.csv', 'a', newline = '',  encoding='utf-8-sig')
wr = csv.writer(f)
roro = 0;


driver = webdriver.Chrome("./chromedriver")

driver.implicitly_wait(3)

driver.get("https://www.dogdrip.net/dogdrip?page=9521")

for i in range(0, 10000):
    for artinum in range(1, 21):

        artixpath = "/html/body/div[1]/main/div/div[2]/div/div[2]/table/tbody/tr[" + str(artinum) + "]/td[1]/span/a/span[1]"
        driver.find_element(By.XPATH, artixpath).click()

        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")

        table = soup.find_all("div","ed margin-bottom-xxsmall margin-left-xsmall", "p")

        for tv in table:
            string = tv.text
            string = string.strip()
            if '@' in string:
                continue
            else:
                wr.writerow([roro, string, ''])
                roro = roro + 1

        driver.back()
    driver.find_element(By.XPATH, "/html/body/div[1]/main/div/div[2]/div/div[2]/div[4]/ul[1]/li[6]/a/span").click()


f.close()
```
